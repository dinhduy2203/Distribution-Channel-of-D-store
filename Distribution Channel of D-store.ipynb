```python
#Đọc dữ liệu
import pandas as pd
df = pd.read_csv('US_Regional_Sales_Data.csv')
df.head()
```

```python
#Data Summary
df.info()
```

```python
#Check for Null Data
df.isnull().sum()
```

```python
#Duplicate Data Verification
duplicate_rows = df.duplicated(keep=False)
print(duplicate_rows)
```

```python
df.head()
```

```python
#Đổi kiểu dữ liệu của các cột: ProcureDate, Orderdate, ShipDate, DeliveryDate
df['ProcuredDate'] = pd.to_datetime(df['ProcuredDate'])
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['ShipDate'] = pd.to_datetime(df['ShipDate'])
df['DeliveryDate'] = pd.to_datetime(df['DeliveryDate'],format = '%Y/%m/%d')
df.info()
df.head()
```

```python
df['DeliveryDate'] = pd.datetime.strptime(df['DeliveryDate'],'%m-%d-%Y')
df.head()
```

```python
#Xử lý cột Unit Cost và Unit Price
df['Unit Cost'] = df['Unit Cost'].str.replace('$',"").str.replace(',','').astype(float)
df['Unit Price'] = df['Unit Price'].str.replace('$',"").str.replace(',','').astype(float)
df.head()
```
